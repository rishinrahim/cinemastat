```python
#imprting all libraries required

import requests
from bs4 import BeautifulSoup
import pandas as pd
import html5lib
import numpy as np
from datetime import datetime
```

```python
FEATURES = ['day','month','year' , 'title', 'director','producer', 'cast', 'genre','story','screenplay','gross','distributor',
            'composer','lyricist','no_of_songs', 'singers','editor','cinematography','banner','dialogue',
            'art_direction','running_time','budget','state_awards','national_awards','plot','imdb_rating',
            'imdb_vote_count']
#function to get the table in wikilink
def get_table(url,idname,classname):

    url = requests.get(url).text
    soup = BeautifulSoup(url,"lxml")
    if idname:
        table = soup.find("table",{"id":idname})
    else:
        table = soup.find("table",{"class":classname})

    return str(table)

#Convert string to date
def convert2date(date_text):
    res = datetime.strptime(date_text, '%b %d %Y').date()
    return res
```

```python
def get_data_wiki(year):
    url ="https://en.wikipedia.org/wiki/List_of_Malayalam_films_of_"+year
    t = get_table(url,None,"wikitable")
    data = pd.read_html(t)[0]
    return data
    
```
