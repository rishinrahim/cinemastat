```python
#ADDING YEAR to the column derived from the movie name.
import re
def get_year(text):
    pattern = re.compile(r"\((\d+)\)")
    year = ''.join(pattern.findall(text))
    try:
        if int(year)>2021:
            return "UNKNOWN"
        else:
            return int(year)
    except:
        return "UNKNOWN"
        

df["year"] = df.movie.apply(get_year)
df.head()
```

```python
#Rearranging columns
ignore = ["background_music",'shooting_location','original_story',"distribution"]
columns =['year','movie', 'movie_genre','director', 'editing','producer','banner','screenplay',
          'music','art_direction','choreography', 'cinematography', 'costumes', 
          'design', 'dialogue','lyrics', 'makeup',  'singers','story','lead_cast','supporting_cast','status']
df = df[columns]
df.shape
```
